```python
import requests

# OIL token - using pair address from DexScreener
pair_address = '8teu1ntubmkunlmw8usnhdwpygcm1md6nvkwc65ldmz2'

print('üõ¢Ô∏è OIL TOKEN TRADING ANALYSIS')
print('='*50)

# Get OIL token data from DexScreener using pair address
response = requests.get(f'https://api.dexscreener.com/latest/dex/pairs/solana/{pair_address}', timeout=10)

if response.status_code == 200:
    data = response.json()
    pair = data.get('pair') or data.get('pairs', [{}])[0] if data.get('pairs') else None
    
    if pair:
        symbol = pair.get('baseToken', {}).get('symbol', 'OIL')
        name = pair.get('baseToken', {}).get('name', 'Unknown')
        mint = pair.get('baseToken', {}).get('address', '')
        price = float(pair.get('priceUsd', 0) or 0)
        liq = float(pair.get('liquidity', {}).get('usd', 0) or 0)
        vol_24h = float(pair.get('volume', {}).get('h24', 0) or 0)
        price_change_24h = float(pair.get('priceChange', {}).get('h24', 0) or 0)

        print(f'Token: {symbol} ({name})')
        print(f'Mint: {mint}')
        print(f'Current Price: ${price:.8f}')
        print(f'Liquidity: ${liq:,.0f}')
        print(f'24h Volume: ${vol_24h:,.0f}')
        print(f'24h Change: {price_change_24h:+.1f}%')

        # Current position
        current_holding = 42264.1077
        current_value = current_holding * price
        print(f'\nüìä CURRENT POSITION:')
        print(f'Holding: {current_holding:,.4f} {symbol}')
        print(f'Current Value: ${current_value:.2f}')

        # Trading decision logic
        should_trade = False
        action = ''
        reason = ''

        if price_change_24h > 25 and vol_24h > 50000:
            should_trade = True
            action = 'BUY MORE'
            reason = f'Strong upward momentum (+{price_change_24h:.1f}%) with high volume'
        elif price_change_24h < -40:
            should_trade = True
            action = 'BUY DIP'
            reason = f'Significant dip ({price_change_24h:.1f}%) - potential recovery'
        elif price_change_24h > 10 and vol_24h > 10000:
            should_trade = True
            action = 'MONITOR'
            reason = f'Moderate momentum (+{price_change_24h:.1f}%) with decent volume'
        else:
            reason = f'Market conditions neutral ({price_change_24h:+.1f}% change, ${vol_24h:,.0f} volume)'

        print(f'\nüí° DECISION: {"‚úÖ " + action if should_trade else "‚ùå HOLD"}')
        print(f'Reason: {reason}')

        if should_trade and action in ['BUY MORE', 'BUY DIP']:
            available_capital = 15.30
            position_size = available_capital * 0.20
            oil_quantity = position_size / price if price > 0 else 0

            print(f'\nüéØ TRADE PROPOSAL:')
            print(f'Action: {action}')
            print(f'Amount: ${position_size:.2f} ({oil_quantity:.0f} {symbol})')
            print(f'Expected Entry: ${price:.8f}')
    else:
        print('No pair data found')
        print(f'Response: {data}')
else:
    print(f'DexScreener API error: {response.status_code}')
```

```python
# Deeper analysis for $5 snipe decision
pair_address = '8teu1ntubmkunlmw8usnhdwpygcm1md6nvkwc65ldmz2'
response = requests.get(f'https://api.dexscreener.com/latest/dex/pairs/solana/{pair_address}', timeout=10)

if response.status_code == 200:
    data = response.json()
    pair = data.get('pair') or data.get('pairs', [{}])[0] if data.get('pairs') else None
    
    if pair:
        price = float(pair.get('priceUsd', 0) or 0)
        liq = float(pair.get('liquidity', {}).get('usd', 0) or 0)
        vol_24h = float(pair.get('volume', {}).get('h24', 0) or 0)
        vol_6h = float(pair.get('volume', {}).get('h6', 0) or 0)
        vol_1h = float(pair.get('volume', {}).get('h1', 0) or 0)
        
        chg_24h = float(pair.get('priceChange', {}).get('h24', 0) or 0)
        chg_6h = float(pair.get('priceChange', {}).get('h6', 0) or 0)
        chg_1h = float(pair.get('priceChange', {}).get('h1', 0) or 0)
        chg_5m = float(pair.get('priceChange', {}).get('m5', 0) or 0)
        
        txns = pair.get('txns', {})
        buys_24h = txns.get('h24', {}).get('buys', 0)
        sells_24h = txns.get('h24', {}).get('sells', 0)
        buys_1h = txns.get('h1', {}).get('buys', 0)
        sells_1h = txns.get('h1', {}).get('sells', 0)
        
        print('üéØ $5 SNIPE ANALYSIS')
        print('='*50)
        print(f'Price: ${price:.8f}')
        print(f'\nüìà PRICE MOMENTUM:')
        print(f'  5m:  {chg_5m:+.2f}%')
        print(f'  1h:  {chg_1h:+.2f}%')
        print(f'  6h:  {chg_6h:+.2f}%')
        print(f'  24h: {chg_24h:+.2f}%')
        
        print(f'\nüìä VOLUME:')
        print(f'  1h:  ${vol_1h:,.0f}')
        print(f'  6h:  ${vol_6h:,.0f}')
        print(f'  24h: ${vol_24h:,.0f}')
        
        print(f'\nüîÑ TRANSACTIONS:')
        print(f'  24h: {buys_24h} buys / {sells_24h} sells')
        print(f'  1h:  {buys_1h} buys / {sells_1h} sells')
        buy_ratio = buys_1h / max(sells_1h, 1)
        
        # Snipe decision
        print(f'\nüí° SNIPE DECISION:')
        snipe_amount = 5.00
        tokens_to_buy = snipe_amount / price if price > 0 else 0
        
        # Check for reversal signals
        reversal = chg_1h > 0 and chg_24h < -20  # Bouncing after big dip
        momentum_up = chg_5m > 0 and chg_1h > 0
        good_volume = vol_1h > 5000
        buy_pressure = buy_ratio > 1.2
        
        signals = []
        if reversal: signals.append('‚úÖ Reversal pattern (1h green after 24h dip)')
        if momentum_up: signals.append('‚úÖ Short-term momentum positive')
        if good_volume: signals.append('‚úÖ Active trading volume')
        if buy_pressure: signals.append(f'‚úÖ Buy pressure ({buy_ratio:.1f}x)')
        
        if not reversal: signals.append('‚ö†Ô∏è No clear reversal yet')
        if not momentum_up: signals.append('‚ö†Ô∏è Still declining short-term')
        if chg_24h < -30: signals.append('‚ö†Ô∏è Major 24h decline (-30%+)')
        
        for s in signals:
            print(f'  {s}')
        
        score = sum([reversal, momentum_up, good_volume, buy_pressure])
        
        print(f'\nüìã SIGNAL SCORE: {score}/4')
        if score >= 3:
            print(f'üöÄ RECOMMENDATION: BUY ${snipe_amount} ({tokens_to_buy:,.0f} OIL)')
            print(f'   Good entry - multiple bullish signals')
        elif score >= 2:
            print(f'ü§î RECOMMENDATION: SMALL BUY ${snipe_amount/2:.2f}')
            print(f'   Mixed signals - reduce position size')
        else:
            print(f'‚è∏Ô∏è RECOMMENDATION: WAIT')
            print(f'   Wait for reversal confirmation')
```

```python
# Search for OIL tokens on DexScreener
search_response = requests.get('https://api.dexscreener.com/latest/dex/search?q=OIL%20solana', timeout=10)
if search_response.status_code == 200:
    search_data = search_response.json()
    pairs = search_data.get('pairs', [])
    print(f'Found {len(pairs)} OIL-related pairs on Solana:')
    for p in pairs[:10]:
        chain = p.get('chainId', '')
        if 'solana' in chain.lower():
            symbol = p.get('baseToken', {}).get('symbol', '?')
            name = p.get('baseToken', {}).get('name', '?')
            address = p.get('baseToken', {}).get('address', '?')
            price = p.get('priceUsd', 0)
            liq = p.get('liquidity', {}).get('usd', 0)
            print(f'  {symbol} ({name}): ${price} | Liq: ${liq:,.0f}')
            print(f'    Address: {address}')
else:
    print(f'Search failed: {search_response.status_code}')
```
